```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('planets')
df.head()
```

# (1) Seleccion de columnas

```python
# Seleccionar una columna
print(df["mass"].head(3)) # selecciono una sola columna
print()

# seleccionar varias columnas a la vez
print(df[["mass", "year"]].head(3)) # selecciono varias columnas como una lista de listas
print()
print(df.iloc[:, [1,3,4]].head(3)) # recomendable si el dataset es pequeño
print()

# Seleccionar una porcion de columnas
print(df.iloc[:, 1:4].head(3)) # con el operador rebanador puedo seleccionar porciones de columnas
```

# (2) seleccionar filas

```python
# seleccionar una fila
print(df.loc[3]) # la funcion loc opera a nivel de indice
print()
print(df.iloc[3]) # la funcion iloc opera a nivel de pocision de filas


# Seleccionar una porcion de filas
print(df.loc[1:5]) # tomo las filas del 1 al 5 inclusive
print()
print(df.iloc[1:5]) # tomo desde la posicion inicial hasta la final-1
print()


# Seleccionar filas no consecutivas
print(df.iloc[[1,5,89]]) # lo hacemos con la funcion iloc, especificando en una lista las fuilas que queremos ver
```

# (3) Seleccionar una porcion del dataframe

```python
# usando la funcion loc
print(df.loc[2:5][["method","mass","year"]]) # aca puedo especificar que columnas quiero ver
print()

# usando la funcion iloc
print(df.iloc[2:5, [0,3,5]]) # No es tan recomendable en datasets demasiado grandes
print()


# variante de iloc
print(df.iloc[12:17, 0:4]) # Selecciono una porcion  de filas
```

# (4) Seleccionar un elemento de mi dataframe

```python
# Usando la funcion loc
print(df.loc[89]["year"]) # recomendable cuando tengo datasets muy grandes

# usando la funcion iloc
print(df.iloc[89,5])
```

# (5) filtros

```python
# filtro 1 (evaluar una sola condicion)

print(df[df["year"] > 2011].head(3))
print()
# filtro 2 (evaluar multiples condiciones, & implica que ambas condiciones deben cumplirse)

print(df[(df["method"] == "Radial Velocity") & (df["year"] > 2011)].head(3))
print()
# filtro 3 (evaluar multiples condiciones, |  implica que al menos una condicion se cumple)

print(df[(df["method"] == "Radial Velocity") | (df["year"] > 2011)].head(3))
print()

# filtro 4 ( negar una condicion)
print(df[~(df["method"] == "Radial Velocity")].head(3)) #~ negara la condicion, por tanto mostrara todos los planetas descubiertos por cualquier metodo excepto radial velocity

# filtro 5
print(df[df["orbital_period"].notna()].head(3)) # mostrara todas las filas que la feature orbital_period no tiene valores nulos
print()

# filtro 6  puedo hacer la busqueda por una cadena de caracteres, por ejempplo, por el primer nombre de una base de datos
print(df[df["method"].str.contains("Radial")].head(3))
print()

# filtro 7 between
print(df[df["mass"].between(1, 10)].head(3))  # variante de cond1 & cond2
print()

# filtro 8 la funcion value_counts() cuenta las apariciones de elementos unicos
print(df["method"].value_counts().head(3))
print()

# filtro 9
# querry: filtra filas de manera similar a sql
print(df.query("year > 2006 and method == 'Radial Velocity'").head(3))
```

# (6) funciones importantes

```python
# Funcion value_counts: cuenta las veces que aparece un elemento unico en una serie de pandas
df["year"].value_counts().head()

# funcion groupby: agrupas poor una categoria y ves la columna mass este caso, a dicha columna le aplicas la operacion que requieras
df.groupby("method")["mass"].agg(["min", "max","mean","max"])

# Agrupar por mas categorias
df.groupby(["method","year"])["mass"].agg(["min", "max","mean","max"])

# Costruir una tabla dinamica con pivot_table
df.pivot_table(values="mass", index="method", columns="year", aggfunc="mean")

```

📌 Función pivot_table: permite construir una tabla dinámica (tipo Excel), donde puedes resumir la información aplicando funciones como mean, sum, count, etc., agrupando por categorías.

🔍 En este ejemplo:

values="mass": queremos calcular la media de la masa (mass).

index="method": agrupamos por el método de descubrimiento (filas).

columns="year": mostramos los años como columnas.

aggfunc="mean": calculamos el promedio de masa para cada combinación de método y año.

📊 Resultado: Una tabla donde cada celda representa el promedio de masa de los planetas descubiertos por un método específico en un año determinado.
